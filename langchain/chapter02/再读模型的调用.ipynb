```python
import time

def call_model():
  print("å¼€å§‹è°ƒç”¨æ¨¡å‹...")
  time.sleep(5)
  print("æ¨¡å‹è°ƒç”¨å®Œæˆ.")

def perform_other_tasks():
  for i in range(5):
    print(f"æ‰§è¡Œå…¶ä»–ä»»åŠ¡{i+1}")
    time.sleep(1)

def main():
  start_time = time.time()
  call_model()
  perform_other_tasks()
  end_time = time.time()
  total_time = end_time - start_time
  return f"æ€»å…±è€—æ—¶:{total_time}ç§’"

main_time = main()
print(main_time)
```

```python
import asyncio
import time

async def async_call(llm):
    await asyncio.sleep(5)
    print("å¼‚æ­¥è°ƒç”¨å®Œæˆ")

async def perform_other_tasks():
    await asyncio.sleep(5)
    print("å…¶ä»–ä»»åŠ¡å®Œæˆ")

async def run_async_tasks():
    start_time = time.time()
    await asyncio.gather(
        async_call(None),
        perform_other_tasks()
    )
    end_time = time.time()
    return f"æ€»å…±è€—æ—¶:{end_time - start_time}ç§’"

result = await run_async_tasks()
print(result)
```
